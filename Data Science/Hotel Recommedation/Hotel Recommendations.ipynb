## Purpose

Using expedia hotel recommendation dataset we attempt to develop a model that will predict which hotel to recommend. The data set is large and requires using a subset as well as handling cylical data.

```python
import opendatasets as od
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import seaborn as sns
import plotly.express as px
import plotly.offline as pyo
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import statsmodels.api as sm
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
```

```python
# Dowloading all the files
od.download("https://www.kaggle.com/c/expedia-hotel-recommendations/data?select=train.csv")
```

```python
# Pulling a random sample of records in the train.csv
p = 0.125
train_data = pd.read_csv("train.csv", skiprows= lambda i: i>0 and random.random() > p)
train_data = train_data[train_data['is_booking'] == True]

# Filling in missing orig_dist data
avg_orig_dist = np.mean(train_data['orig_destination_distance'])
train_data.loc[train_data['orig_destination_distance'].isnull(), ['orig_destination_distance']] = avg_orig_dist

# Converting columns into datetime format
train_data['date_time'] = pd.to_datetime(train_data['date_time'], errors = 'coerce')
train_data['srch_ci'] = pd.to_datetime(train_data['srch_ci'], format='%Y-%m-%d %H:%M', errors = 'coerce')
train_data['srch_co'] = pd.to_datetime(train_data['srch_co'], format='%Y-%m-%d %H:%M', errors = 'coerce')


# Dropping records missing check in and check out times
train_data.dropna(inplace=True)

# Splitting datetime data into components and creating sin/cos verision to capture the cyclical aspect of the data
train_data['date_time_year'] = train_data['date_time'].dt.year

train_data['date_time_month'] = train_data['date_time'].dt.month
train_data['date_time_month_sin']=np.sin(train_data['date_time_month']*(2*np.pi/12))
train_data['date_time_month_cos']=np.cos(train_data['date_time_month']*(2*np.pi/12))

train_data['date_time_day'] = train_data['date_time'].dt.day
train_data['date_time_day_sin']=np.sin(train_data['date_time_day']*(2*np.pi/30))
train_data['date_time_day_cos']=np.cos(train_data['date_time_day']*(2*np.pi/30))

train_data['date_time_weekday'] = train_data['date_time'].dt.weekday
train_data['date_time_weekday_sin']=np.sin(train_data['date_time_weekday']*(2*np.pi/7))
train_data['date_time_weekday_cos']=np.cos(train_data['date_time_weekday']*(2*np.pi/7))

train_data['date_time_hour'] = train_data['date_time'].dt.hour
train_data['date_time_hour_sin']=np.sin(train_data['date_time_hour']*(2*np.pi/24))
train_data['date_time_hour_cos']=np.cos(train_data['date_time_hour']*(2*np.pi/24))

train_data['srch_ci_year'] = train_data['srch_ci'].dt.year

train_data['srch_ci_month'] = train_data['srch_ci'].dt.month
train_data['srch_ci_month_sin']=np.sin(train_data['srch_ci_month']*(2*np.pi/12))
train_data['srch_ci_month_cos']=np.cos(train_data['srch_ci_month']*(2*np.pi/12))

train_data['srch_ci_day'] = train_data['srch_ci'].dt.day
train_data['srch_ci_day_sin']=np.sin(train_data['srch_ci_day']*(2*np.pi/30))
train_data['srch_ci_day_cos']=np.cos(train_data['srch_ci_day']*(2*np.pi/30))

train_data['srch_co_year'] = train_data['srch_co'].dt.year

train_data['srch_co_month'] = train_data['srch_co'].dt.month
train_data['srch_co_month_sin']=np.sin(train_data['srch_co_month']*(2*np.pi/12))
train_data['srch_co_month_cos']=np.cos(train_data['srch_co_month']*(2*np.pi/12))

train_data['srch_co_day'] = train_data['srch_co'].dt.day
train_data['srch_co_day_sin']=np.sin(train_data['srch_co_day']*(2*np.pi/30))
train_data['srch_co_day_cos']=np.cos(train_data['srch_co_day']*(2*np.pi/30))
```

```python
# Splitting the data into a training and testing set
train_df, test_df = train_test_split(train_data, test_size=0.3)
x_train = train_df.loc[:,train_df.columns != 'hotel_cluster']    #target column
y_train = train_df[['hotel_cluster']] 
x_test = test_df.loc[:,test_df.columns != 'hotel_cluster']  #independent columns
y_test = test_df[['hotel_cluster']]    #target column


x_train = x_train.loc[:,x_train.columns != 'date_time']
x_train = x_train.loc[:,x_train.columns != 'srch_ci']
x_train = x_train.loc[:,x_train.columns != 'srch_co']
x_test = x_test.loc[:,x_test.columns != 'date_time']
x_test = x_test.loc[:,x_test.columns != 'srch_ci']
x_test = x_test.loc[:,x_test.columns != 'srch_co']
```

```python
# Creating a correlation heatmap of the variables
import numpy as np
# Compute the correlation matrix
corr = x_train.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)


# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}).set_title('Heatmap Correlation Matrix')
plt.show()
```

```python
#Backward Elimination
cols = list(x_train.columns)
pmax = 1
while (len(cols)>0):
    p= []
    model = sm.GLS(y_train.astype('float'),x_train.astype('float')).fit()
    p = pd.Series(model.pvalues,index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
x_train = x_train[['user_location_country', 'orig_destination_distance', 'user_id', 'is_mobile', 'is_package', 'channel', 
                   'srch_adults_cnt', 'srch_children_cnt', 'srch_rm_cnt', 'srch_destination_id', 'srch_destination_type_id',
                   'hotel_continent', 'hotel_market', 'srch_ci_year', 'srch_ci_month', 'srch_ci_day', 'srch_co_year', 
                   'srch_co_month', 'srch_co_day']]

                  
x_test = x_test[['user_location_country', 'orig_destination_distance', 'user_id', 'is_mobile', 'is_package', 'channel', 
                   'srch_adults_cnt', 'srch_children_cnt', 'srch_rm_cnt', 'srch_destination_id', 'srch_destination_type_id',
                   'hotel_continent', 'hotel_market', 'srch_ci_year', 'srch_ci_month', 'srch_ci_day', 'srch_co_year', 
                   'srch_co_month', 'srch_co_day']]
```

```python
rfc_model = RandomForestClassifier()
rfc_model.fit(x_train,y_train)

predicted = rfc_model.predict(x_test)
report = classification_report(y_test, predicted)

print(report)
print('Confusion Matrix:')
confusion_matrix(y_test,predicted)
```

```python
# Model Test - Naive Bayesian Classifier
# Then using classification_report to see precision, recall, and 
# f1-score for each possible outcome, and overall average.

nbc_model = GaussianNB()
nbc_model.fit(x_train,y_train)

predicted = nbc_model.predict(x_test)
report = classification_report(y_test, predicted)

print(report)
print('Confusion Matrix:')
confusion_matrix(y_test,predicted)
```

```python
# Model Test - Decision Tree Classifier
# Then using classification_report to see precision, recall, and 
# f1-score for each possible outcome, and overall average.

dtc_model = DecisionTreeClassifier()
dtc_model.fit(x_train,y_train)

predicted = dtc_model.predict(x_test)
report = classification_report(y_test, predicted)

print(report)
print('Confusion Matrix:')
confusion_matrix(y_test,predicted)
```

## Conclusion
The data was pulled by taking a sample of the train.csv data from the Expedia dataset hosted on Kaggle https://www.kaggle.com/c/expedia-hotel-recommendations. As the purpose of the project was to predict the hotel_cluster it was decided to filter off the data for records that only have bookings. As booking a hotel typically fluctuates during specific season, or periods, the project attempted capture this cyclical nature by decomposing the date time variables into their components, i.e year, month, day, etc. From there to capture the cyclical aspect a sine and cosine function was applied to the fields. 

A backwards feature elimination was performed to come up with a final set of features to model with. Using these features three models were created using the following classifiers: Random Forest, Naive Bayes, and Decision Tree. All three performed poorly. With the Naive Bayes performing the worst at 6% accuracy, then Decision Tree at 12%, and Random Forest at 13%.

During this project some technical issues did come up. Primarily with the memory performance of the laptop. There was limit to how much data was able to be pulled in. This lead to long run times and kernel failings. With the ability to pull in more data, the class imbalance we in the testing data set would be minimized.
