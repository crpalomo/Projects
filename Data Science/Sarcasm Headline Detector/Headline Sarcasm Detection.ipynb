```python
# Import libraries
import numpy as np
import pandas as pd
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Import neccesary machine learning libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score,recall_score
from sklearn import svm
from scipy.sparse import hstack

warnings.filterwarnings("ignore")
```

```python
# Load Seaborn color palete
palette = sns.color_palette("colorblind")
#Load analyzers
sid = SentimentIntensityAnalyzer()
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

#Load stopwords
stopwords = nltk.corpus.stopwords.words('english')
```

```python
#Load data
headlines_df = pd.concat([pd.read_json('data/Sarcasm_Headlines_Dataset.json',lines=True),
                          pd.read_json('data/Sarcasm_Headlines_Dataset_v2.json',lines=True)])
headlines_df.drop(columns=['article_link'], inplace=True)
headlines_df.reset_index(drop=True,inplace=True)

headlines_df['headlines_without_stopwords'] = headlines_df['headline'].apply(
    lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
headlines_df['word_count'] = headlines_df['headline'].apply(lambda x: len(x.split(' ')))
```

```python
#Function to add sentiment analysis columns based on a source column
def sentiment_analysis_addition(df, source_column):
    for index, row in df.iterrows():
        df.loc[index,f'{source_column}_vader_compound_score'] = round(sid.polarity_scores(row[f'{source_column}'])['compound'],2)
        doc = nlp(row[f'{source_column}'])
        df.loc[index,f'{source_column}_scapy_polarity'] = round(doc._.blob.polarity,2)
        df.loc[index,f'{source_column}_scapy_subjectivity'] = round(doc._.blob.subjectivity,2)
    return df
```

```python
# Run function to sentiment analysis data
headlines_df = sentiment_analysis_addition(headlines_df,'headline')
```

```python
# Create test and training data set

X = headlines_df.drop(columns=['is_sarcastic','headline'])
y = headlines_df.is_sarcastic
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
cv = CountVectorizer(ngram_range=(1,3))
X_cv_train = cv.fit_transform(X_train['headlines_without_stopwords'])
X_cv_test = cv.transform(X_test['headlines_without_stopwords'])

numerical_columns = ['word_count','headline_vader_compound_score', 'headline_scapy_polarity',
                     'headline_scapy_subjectivity']
for nc in numerical_columns:
    X_cv_train= hstack((X_cv_train,np.array(X_train[f'{nc}'])[:,None])).tocsr()
    X_cv_test= hstack((X_cv_test,np.array(X_test[f'{nc}'])[:,None])).tocsr()
```

## Model Creation and Scoring

```python
# Logestic Regression Model
model = LogisticRegression()
model.fit(X_cv_train, y_train)
model.score(X_cv_train, y_train)
y_pred = model.predict(X_cv_test)
logreg_accuracy = accuracy_score(y_test,y_pred)
logreg_f1 = f1_score(y_test,y_pred)
logreg_precision = precision_score(y_test,y_pred)
logreg_recall = recall_score(y_test,y_pred)
```

```python
# Random Forest Model
model = RandomForestClassifier(criterion='gini',
                                n_estimators=12, 
                                random_state=1)

model.fit(X_cv_train, y_train)
model.score(X_cv_train, y_train)
y_pred = model.predict(X_cv_test)
randforest_accuracy = accuracy_score(y_test,y_pred)
randforest_f1 = f1_score(y_test,y_pred)
randforest_precision = precision_score(y_test,y_pred)
randforest_recall = recall_score(y_test,y_pred)
```

```python
# K Nearest Neighbor Model
model = KNeighborsClassifier(n_neighbors = 1)

model.fit(X_cv_train, y_train)
model.score(X_cv_train, y_train)
y_pred = model.predict(X_cv_test)
knn_accuracy = accuracy_score(y_test,y_pred)
knn_f1 = f1_score(y_test,y_pred)
knn_precision = precision_score(y_test,y_pred)
knn_recall = recall_score(y_test,y_pred)
```

```python
# SVM Model
model = svm.SVC(kernel='linear', gamma=1) 
model.fit(X_cv_train, y_train)
model.score(X_cv_train, y_train)
y_pred = model.predict(X_cv_test)
svm_accuracy = accuracy_score(y_test,y_pred)
svm_f1 = f1_score(y_test,y_pred)
svm_precision = precision_score(y_test,y_pred)
svm_recall = recall_score(y_test,y_pred)
```

```python
# Create dataframe of results

models = ['LogisticReg', 'RandomForest', 'KNN', 'SVM']
acc = [logreg_accuracy,randforest_accuracy, knn_accuracy,svm_accuracy ]
f1s = [logreg_f1, randforest_f1, knn_f1, svm_f1]
pre = [logreg_precision,randforest_precision, knn_precision, svm_precision ]
rec = [logreg_recall,randforest_recall, knn_recall, svm_recall]

results = {'Model': models, 'Accuracy':acc, 'F1-Score':f1s, 'Precision': pre, 'Recall':rec}
results_df = pd.DataFrame(results)
results_df.head()
```
