## Purpose

Using data from doger attendance determine which promotion and day combination will most likely increase the amount of attendees.

```python
# Loading Libraries
import pandas as pd
import plotly.express as px
import plotly.offline as pyo
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import statsmodels.api as sm
import numpy as np
import kaleido
import plotly.io as pio
pio.kaleido.scope.default_format = "svg"
```

```python
# Loading data into dataframe
dodgers_df = pd.read_csv('dodgers.csv')

# Creating a column to denote what type promotion, if any, happened that day. This is to deal with seperate columns
# of YES's and NO's for each promotion.

promotion = ['cap', 'shirt', 'fireworks', 'bobblehead']
dodgers_df['promotion'] ='none'
for p in promotion:
    dodgers_df.loc[(dodgers_df[f'{p}'] == 'YES') &
                   (dodgers_df['promotion'] != 'none'), 'promotion'] = dodgers_df['promotion'] + f', {p}'
    dodgers_df.loc[(dodgers_df[f'{p}'] == 'YES') &
                   (dodgers_df['promotion'] == 'none'), 'promotion'] = f'{p}'

# Creating a Month/Day columns, in order to view data chornologically
dodgers_df['date'] = dodgers_df['month'] + '/'+ dodgers_df['day'].astype(str) + '/2012'
dodgers_df['date'] = pd.to_datetime(dodgers_df['date'])

# Print dodgers_df
dodgers_df.head()
```

## Exploratory Visualization
To begin with I will preform exploaratory visualization to understand the data better. My initial thought is that the various promotions will have different effect, and that the day of the week will play role. Looking at the visualization we can see that there were two cap promotions and three shirt promotions. Also with the exception of the Fourth of July every firework promotion was on Friday. ## Regression model

```python
# Using the the .get_dummies module from pandas to encode the categorical variables. Dropping the originial columns
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res) 

model_df = dodgers_df.copy()
categories = ['month', 'day_of_week', 'opponent', 'skies', 'day_night', 'cap', 'shirt', 'fireworks', 'bobblehead']

for c in categories:
    model_df = encode_and_bind(model_df,f'{c}')

# Removing the created promotion and date columns
model_df = model_df.drop(columns = ['promotion', 'date'])


model_df.head()
```

```python
# Seperating data frame to dataframes. the "x" is the feature, or categories, that I will be investigating
# The "y" is the results, which is the attendence
x = model_df.loc[:, model_df.columns != 'attend']
y = model_df.loc[:, model_df.columns == 'attend']

#Backward Elimination
cols = list(x.columns)
pmax = 1
while (len(cols)>0):
    p= []
    model = sm.GLS(y,x).fit()
    p = pd.Series(model.pvalues,index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
# Updating model dataframe to use only the resulting caolumns from the backwards elimination

model_df = model_df[['day_of_week_Friday', 'day_of_week_Saturday', 'day_of_week_Tuesday', 'skies_Clear ',
                     'skies_Cloudy', 'day_night_Day', 'fireworks_YES', 'bobblehead_YES','attend']]

# Updating x, and y dataframe
x = model_df.loc[:, model_df.columns != 'attend']
y = model_df.loc[:, model_df.columns == 'attend']

# Creating a model
model = sm.GLS(y,x)

# Fitting the model
result = model.fit()

# Printing summary of the model
result.summary()
```

```python
# Repeating previous steps to develop a generalized linear model
# Using the the .get_dummies module from pandas to encode the categorical variables. Dropping the originial columns
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res) 

model_df = dodgers_df.copy()
categories = ['month', 'day_of_week', 'opponent', 'skies', 'day_night', 'cap', 'shirt', 'fireworks', 'bobblehead']

for c in categories:
    model_df = encode_and_bind(model_df,f'{c}')

# Removing the created promotion and date columns
model_df = model_df.drop(columns = ['promotion', 'date'])


model_df.head()

# Seperating data frame to dataframes. the "x" is the feature, or categories, that I will be investigating
# The "y" is the results, which is the attendence
x = model_df.loc[:, model_df.columns != 'attend']
y = model_df.loc[:, model_df.columns == 'attend']

#Backward Elimination
cols = list(x.columns)
pmax = 1
while (len(cols)>0):
    p= []
    model = sm.GLM(y,x).fit()
    p = pd.Series(model.pvalues,index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
# Updating model dataframe to use only the resulting caolumns from the backwards elimination

model_df = model_df[['month_AUG', 'day_of_week_Friday', 'day_of_week_Saturday', 'day_of_week_Tuesday',
                     'skies_Clear ', 'skies_Cloudy', 'day_night_Day', 'cap_NO', 'shirt_YES', 'fireworks_YES',
                     'bobblehead_YES','attend']]

# Updating x, and y dataframe
x = model_df.loc[:, model_df.columns != 'attend']
y = model_df.loc[:, model_df.columns == 'attend']


# Creating a Generalized Linear model
model = sm.GLM(y,x)

# Fitting the model
result = model.fit()

# Printing summary of the model
result.summary()
```

## Conclusion

My initial thoughts when looking at the data was that the day of the week and promotion type would play the effect. I believed this as the ability to go the game dependent on schedules, so there was an expectation that the amount of the attended during the work week would be less than the weekend. The promotions are an added incentive therefore would have different repones as a driving force to encourage attendance. The feature that I believed would have the least role would be the day. As the day of the week for each day varies between months and years, and there is no guaranteeing that would be a date for a specific day. 

To begin this assignment, I created two columns a date based of the month, day, and the year being 2012, and promotions columns that showed which promotion was applied that day. These two were help with exploratory visualization aspect. From the graphs I noticed that Friday has the greatest amount of total attendance of any weekday and every Friday was fireworks promotion. From the box plot I was able to see that there were no outliers in terms of promotion and attendance. Though through a violin graph I was able to determine that the box plot spread of the cap and shirts promotion was somewhat misleading as there only done 2 and 3 time respectively. This indicated that the model might fully represent the true effect that this promotion can have.

To create the model, I encoded the categorical variables 'month', 'day_of_week', 'opponent', 'skies', 'day_night', 'cap', 'shirt', 'fireworks', and 'bobblehead' and separated out their level into their own columns. With 0 and 1 to denote if that level true for that record. From there I used Backwards elimination to filter out for the needed features. Script would create a general least square regression model, then remove any of factors that had greater than 0.05 p-value. It would repeat this process until there the final list features were found.

As a means of validating the results, and to get an understanding if the residuals are normally distributed, which is an assumption of the general least square regression model, I created a generalized linear regression model. Comparing the results, I found that the key features were similar. The most notable difference being that the generalized linear model considered if hats were given as a key feature. Based on of these results would propose the following.

Offer a bobblehead promotions on Saturday and Thursdays, especially on clear days. The reason for this is that based off the model Saturday, clear days, and bobbleheads have a positive effect on attendance. So, this should increase attendance on Saturday. Though Thursday was not in the model, looking at the historical data we can see that Thursday has the lowest sales across the days of the week. In fact, the two bobblehead promotion days performed better than the non-promotion days. So, taking the fact that clear days and bobbleheads should increase attendance according to the model. This should increase attendance on the lowest preforming day.
